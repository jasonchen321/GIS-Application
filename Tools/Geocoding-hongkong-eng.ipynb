```python
addr = []
for i in range(len(add)):
    for j in add[i]:
        addr.append(j)
```

```python
District=[]
Street_name=[]
House_number=[]
Lat=[]
Lon=[]
for i in addr:
    response = requests.get("https://api.tomtom.com/search/2/geocode/"+i+".json?countrySet=HK&lat=37.337&lon=-121.89&topLeft=37.553%2C-122.453&btmRight=37.4%2C-122.55&language=zh-TW&key=UJQpPv18GXBXaN4J3W5myUazZtTdDzG8")
    response = json.loads(response.text)
    print(i)
    try:
        x = response['results'][0]['position']['lon']
        y = response['results'][0]['position']['lat']
        street = response['results'][0]['address']['streetName']
        House = response['results'][0]['address']['streetNumber']
        dist = response['results'][0]['address']['localName']
    except Exception as e:
        if 'local' in str(e):
            dist = ""
        elif 'Number' in str(e):
            House = ""
        elif 'Name' in str(e):
            street = ""
        else:
            dist = ""
            street = ""
            House = ""
            x = ""
            y = ""
            
    District.append(dist)
    Street_name.append(street)
    House_number.append(House)
    Lat.append(y)
    Lon.append(x)
    print(i)
```
